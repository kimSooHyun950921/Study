## 05-1 결정트리

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
wine = pd.read_csv('https://bit.ly/wine_csv_data')
wine.info()
```

```python
wine.describe()
train = wine[['alcohol','sugar','pH']]
test = wine[['class']]
```

```python
train_input, test_input, train_target, test_target = \
    train_test_split(train, test, random_state= 42)
```

```python
dc = DecisionTreeClassifier(random_state=42)
dc.fit(train_input, train_target)
dc.score(train_input, train_target)
```

```python
dc.score(test_input, test_target)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(10, 7))
plot_tree(dc, max_depth=1, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```

```python
dt = DecisionTreeClassifier(min_impurity_decrease=0.0005, max_depth=3)
dt.fit(train_input, train_target)
dt.score(test_input, test_target)
```

```python
plt.figure(figsize=(10, 7))
plot_tree(dt, max_depth=3, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```
